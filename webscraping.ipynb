### Libaries, functions, constants

```python
import requests
from bs4 import BeautifulSoup
import unicodedata
import mysql.connector

def boil_soup(spice, elixir):
    soup = BeautifulSoup(spice, elixir)
    return soup

def cleanLists(thisList):
    newList = []
    for item in thisList:
        text = str(item.get_text(strip=True) )
        clean_text = unicodedata.normalize("NFKD", text)
        if clean_text:
            newList.append(clean_text)
    return newList

def create_tables(tablename):
    create_table = '''CREATE TABLE {} ( ValueList VARCHAR(1024) )'''.format(tablename)

    try:
        db.execute('DROP TABLE if exists {}'.format(tablename))
        db.execute(create_table)
        print(tablename, "table created")
    except Exception as err:
        print("DB error:", err)
        
def insertRows(tablename, data):
    add = '''INSERT INTO {} ( ValueList ) VALUES (%s)'''.format(tablename)
    try:
        db.execute(add, [data])
        cn.commit()
        return True;
    except mysql.connector.Error as InsertError:
        print("Failed to insert record into MySQL table: {}".format(InsertError))
        return False;
    except Exception as err:
        print("Server error:", err)
        return False;

#constants used
site = "https://www.straitstimes.com/global"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Brave Chrome/83.0.4103.116 Safari/537.36'}
tablenames = ["STHeadlines", "STSubHeadlines", "Header1", "links"]
```

### Identify the tag which has the headlines and sub headlines and extract them using beautiful soup. (10 Marks)
